# Train Test Split

```python
X=concrete.drop('Strength',axis=1)
y=concrete['Strength']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Plotting the Distribution without any Transformation

```python
sns.displot(concrete['Cement'])
```

```python
sns.distplot(concrete['Cement'])
```

```python
for col in X_train.columns:
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)
    plt.show()
    
    plt.figure(figsize=(10,10))
    plt.subplot(122)
    stats.probplot(X_train[col],dist='norm',plot=plt)
    plt.title(col)
    plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
```

```python
X_train_transformed=pt.fit_transform(X_train)
X_test_transformed=pt.transform(X_test)
```

```python
X_train_transformed
```

# Again Distribution

```python
# input_features (input)
X_train_transformed=pd.DataFrame(X_train_transformed,columns=X_train.columns)

for col in X_train_transformed.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)
    
# QQ plot
    plt.subplot(122)
    sns.distplot(X_train_transformed[col])
    plt.title(col)
    plt.show()
```

# Scaling(Standard Scalar)

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_transformed=scaler.fit_transform(X_train_transformed)
X_test_transformed=scaler.fit_transform(X_test_transformed)
```

```python
X_train_transformed
```

# Training Models

pip install xgboost

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRFRegressor

from sklearn.metrics import mean_squared_error,r2_score
```

```python
models={
    "lr":LinearRegression(),
    "lss":Lasso(),
    "Ridg":Ridge(),
    "dtr":DecisionTreeRegressor(),
    "xgb":XGBRFRegressor()
}

for name,md in models.items():
    md.fit(X_train_transformed,y_train)
    y_pred=md.predict(X_test_transformed)
    
    print(f"{name} :mse: {mean_squared_error(y_test,y_pred)}, r2 score:{r2_score(y_test,y_pred)}")
```

# Selecting Best Model

```python
xgb=XGBRFRegressor()
xgb.fit(X_train_transformed,y_train)
y_pred=xgb.predict(X_test_transformed)
r2_score(y_test,y_pred)
```

# Prediction Model

```python
def pred_strength(Cement,BlastFurnace,FlyAsh,Water,Superplasticizer,CoarseAggregate,FineAggregate,Age):
    features=np.array([[Cement,BlastFurnace,FlyAsh,Water,Superplasticizer,CoarseAggregate,FineAggregate,Age]])
    
    pred=xgb.predict(features)
    
    return pred[0]
```

```python
Cement=158.60
BlastFurnace=148.90
FlyAsh=116.00
Water=175.10
Superplasticizer=15.00
CoarseAggregate=953.3
FineAggregate=719.70
Age=28

strength=pred_strength(Cement,BlastFurnace,FlyAsh,Water,Superplasticizer,CoarseAggregate,FineAggregate,Age)
```
